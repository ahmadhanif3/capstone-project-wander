```python
def get_var_name(var):
  for name, value in globals().items():
    if value is var:
      return name
```

```python
# search_string for what data you want to gather
search_string = "Tourist Attraction"   #input('Search string (e.g., Hotel): ')
# cities = daftar_kabupaten
# districts = daftar_kabupaten[city]

# for districts in kalimantan_utara:
#     for city in districts:
#         if city in Kota_Tarakan:  # Check if the city is in Kota_Tarakan
#             for district in city:
#                 folder_path = f"/content/Kota_Tarakan_{district.replace(' ', '_')}"
#                 if not os.path.exists(folder_path):
#                     os.makedirs(folder_path)
#                 print(f"Created folder: {folder_path}")

for kabkot_name, kabkot_value in provinsi.items():
  for allcamat in kabkot_value: # <-- problemnya di allcamat, ini balikin array bkn kota, jd rusak di query
    for camat in allcamat:
      allcamatname = get_var_name(allcamat)
      folder_path = f"/content/{kabkot_name}/{allcamatname}"
      if not os.path.exists(folder_path):
        os.makedirs(folder_path)

      print(f"Size folder: {len(allcamat)} file(s) kecamatan.")

      address = f"{camat}, {allcamatname}" # input('Input address of place or city (e.g., Depok): ')
      business_list = []


      # 1. Using text search to find places within the city
      # 2. element array camats diambil ke variable camat.
      # 3. Disini query = Hotel in Beji, Depok. Apabila input city = 'Depok' & element array yg sedang diambil itu 'Beji'.
      response = map_client.places(query=f"{search_string} in {address}") # <-- ini query rusak karena input, jd ga ngisi business_list


      business_list.extend(response.get('results'))
      next_page_token = response.get('next_page_token')

      while next_page_token:
          time.sleep(2)
          response = map_client.places(query=f"{search_string} in {address}", page_token=next_page_token)
          business_list.extend(response.get('results'))
          next_page_token = response.get('next_page_token')

      # Check for duplicates before adding to the dataframe
      seen_place_ids = set()
      unique_business_list = []
      for business in business_list:
          if business['place_id'] not in seen_place_ids:
              unique_business_list.append(business)
              seen_place_ids.add(business['place_id'])

      df = pd.DataFrame(unique_business_list)
      df['url'] = 'https://www.google.com/maps/place/?q=place_id:' + df['place_id']
      file_name = f"{kabkot_name}_{allcamatname}_{camat}.csv" #input('Input file-name for <file-name>.csv (no whitespace, only - & _): ')

      file_path = os.path.join(folder_path, file_name)
      df.to_csv(file_path, index=False)


      print(f'Exported as {file_name}')
      #files.download(f'/content/{file_name}.csv')
```
